```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import math
from scipy.integrate import odeint
import pandas as pd
import seaborn
from scipy import stats
```

```python
df = pd.read_csv('GroupByBFGs.csv')
```

```python
num=5
WOBFG_Plas=np.zeros(num+1)
WBFG_Plas=np.zeros(num+1)
```

```python
for i in range(np.size(df,0)):
    if df.biofilm_chromosome[i]+df.biofilm_plasmid[i]==0:
        if df.plasmid_count[i]<num:
            WOBFG_Plas[df.plasmid_count[i]]=WOBFG_Plas[df.plasmid_count[i]]+1
        else:
            WOBFG_Plas[-1]=WOBFG_Plas[-1]+1
    else:
        if df.plasmid_count[i]<num:
            WBFG_Plas[df.plasmid_count[i]]=WBFG_Plas[df.plasmid_count[i]]+1
        else:
            WBFG_Plas[-1]=WBFG_Plas[-1]+1
```

```python
plt.figure(figsize=(2,2))
plt.pie(WOBFG_Plas)
plt.savefig('PlasNumDistribution_1.pdf')
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.pie(WBFG_Plas)
plt.savefig('PlasNumDistribution_2.pdf')
plt.show()
```
