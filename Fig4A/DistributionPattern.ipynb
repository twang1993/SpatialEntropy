```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num=700
plt.figure(figsize=(3,3))
xd=np.random.rand(num)
yd=np.random.rand(num)
plt.plot(xd,yd,'o',markersize=3,color='#97BDD4')
xr=np.random.rand(num)
yr=np.random.rand(num)
plt.plot(xr,yr,'o',markersize=3,color='#CE594F')
plt.tick_params(direction='in',top='true',right='true')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xticks([])
plt.yticks([])
plt.savefig('DistributionPattern_1_1.pdf')
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.hist2d(np.concatenate((xd,xr)), np.concatenate((yd,yr)), bins=30,range=[[0,1],[0,1]], cmap='OrRd')
#plt.colorbar()
plt.tick_params(direction='in',top='true',right='true')
plt.clim(0,15)
plt.xticks([])
plt.yticks([])
plt.savefig('DistributionPattern_1_2.pdf')
plt.show()
```

```python
plt.figure(figsize=(3,3))
sss=np.random.rand(num)
ddd=0.2
xd=np.random.normal(loc=0.8,scale=ddd,size=num)*(sss>0.5)+np.random.normal(loc=0.2,scale=ddd,size=num)*(sss<0.5)
yd=np.random.normal(loc=0.8,scale=ddd,size=num)*(sss>0.5)+np.random.normal(loc=0.2,scale=ddd,size=num)*(sss<0.5)
plt.plot(abs(xd),abs(yd),'o',markersize=3,color='#97BDD4')
xr=np.random.normal(loc=0.8,scale=ddd,size=num)*(sss>0.5)+np.random.normal(loc=0.2,scale=ddd,size=num)*(sss<0.5)
yr=np.random.normal(loc=0.8,scale=ddd,size=num)*(sss>0.5)+np.random.normal(loc=0.2,scale=ddd,size=num)*(sss<0.5)
plt.plot(abs(xr),abs(yr),'o',markersize=3,color='#CE594F')
plt.tick_params(direction='in',top='true',right='true')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xticks([])
plt.yticks([])
plt.savefig('DistributionPattern_2_1.pdf')
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.hist2d(np.concatenate((abs(xd),abs(xr))), np.concatenate((abs(yd),abs(yr))), bins=30,range=[[0,1],[0,1]],cmap='OrRd')
#plt.colorbar()
plt.clim(0,15)
plt.xlim([0,1])
plt.ylim([0,1])
plt.tick_params(direction='in',top='true',right='true')
plt.xticks([])
plt.yticks([])
plt.savefig('DistributionPattern_2_2.pdf')
plt.show()
```
